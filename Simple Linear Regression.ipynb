```python
import seaborn as sbrn
import matplotlib.pyplot as mat
def heatmap(df):
  mat.figure(figsize=(6,4))
  sbrn.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
  return mat.show()

heatmap(df)
```

Step 3: Pre-Processing



```python
#deleting the records having NaN values
'''
df=df.dropna()'''
#replacing the values of column with median of values

median_years=df['YearsExperience'].median()
```

```python
median_years
```

```python
mean_sal=df['Salary'].mean()
```

```python
mean_sal
```

```python
import math
mean_sal=math.floor(df['Salary'].mean())
print(mean_sal)
```

```python
df['YearsExperience']=df['YearsExperience'].fillna(median_years)
df['Salary']=df['Salary'].fillna(mean_sal)
```

```python
import seaborn as sbrn
import matplotlib.pyplot as mat
def heatmap(df):
  mat.figure(figsize=(6,4))
  sbrn.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
  return mat.show()

heatmap(df)
```

Step 4:Prepare Data


```python
'''X=df['YearsExperience']
Y=df['Salary']'''
#2d values
X=df.iloc[:,:-1].values
Y=df.iloc[:,1:].values
print(X)
```

Step 5 - Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

Step 6:Training


```python
'''Linear regression'''
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

Fitting the data

```python
model.fit(X_train,Y_train)
```

Step 7-Testing the model

```python
y_predict=model.predict(X_test)
```

Step 8- Model Evaluation 

```python
import matplotlib.pyplot as plt
plt.scatter(X_test,Y_test,color='red')
plt.plot(X_test,y_predict,color='blue')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('ML Model Exam Results')
plt.show()
```

```python
model.score(X_test,Y_test)*100
```

Deployment

```python
import math as m
years=float(input("Enter Years of Experience of the Employee:"))
y_res=model.predict([[years]])
print("Salary estimated by model:",m.floor(y_res))
```
